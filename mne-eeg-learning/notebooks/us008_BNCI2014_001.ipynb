````python

# ============================================================
# 2. 使用 Paradigm 提取特征（epoch）
# ============================================================
# MotorImagery paradigm 会自动完成：
#   - 读取原始 Raw
#   - 带通滤波（默认 4-38 Hz）
#   - 根据事件标记切分 epoch
#   - 返回 X (n_trials, n_channels, n_samples) 和 y (标签)

paradigm = MotorImagery(
    fmin=4.0,       # 高通截止频率
    fmax=38.0,      # 低通截止频率
    interval=(0, 4) # epoch 时间窗：事件后 0~4 秒（可选 0~3）
)

# 以被试 1 为例
subject_id = [1]
X, y, metadata = paradigm.get_data(dataset=dataset, subjects=subject_id)

print("=" * 50)
print("Epoch 数据")
print("=" * 50)
print(f"X shape : {X.shape}")      # (n_trials, n_channels, n_samples)
print(f"y shape : {y.shape}")      # (n_trials,)
print(f"标签类别: {np.unique(y)}") # ['feet','left_hand','right_hand','tongue']
print()
print("metadata 前 5 行:")
print(metadata.head())
print()

# ============================================================
# 3. 查看各类别样本数量
# ============================================================
print("=" * 50)
print("类别分布")
print("=" * 50)
for label in np.unique(y):
    count = np.sum(y == label)
    print(f"  {label:12s} : {count} trials")
print()

# ============================================================
# 4. 加载多个被试
# ============================================================
subject_ids = [1, 2, 3]
X_multi, y_multi, meta_multi = paradigm.get_data(dataset=dataset, subjects=subject_ids)

print("=" * 50)
print("多被试数据")
print("=" * 50)
print(f"被试 {subject_ids} 总样本数: {X_multi.shape[0]}")
for sid in subject_ids:
    mask = meta_multi['subject'] == sid
    print(f"  Subject {sid} : {mask.sum()} trials")
print()

# ============================================================
# 5. 简单可视化（可选）
# ============================================================
import matplotlib.pyplot as plt

# 5a. 画第一个 trial 各通道的时域信号
fig, axes = plt.subplots(4, 1, figsize=(12, 8), sharex=True)
fig.suptitle(f"Subject 1 — Trial 0 — Label: {y[0]}", fontsize=14)

sfreq = 256
n_samples = X.shape[2]
t = np.arange(n_samples) / sfreq

channels_to_plot = [0, 1, 2, 3]  # 前 4 个通道
channel_names = dataset.datasets[0].raw.ch_names

for i, ch_idx in enumerate(channels_to_plot):
    axes[i].plot(t, X[0, ch_idx, :], linewidth=0.8, color='steelblue')
    axes[i].set_ylabel(channel_names[ch_idx])
    axes[i].grid(True, alpha=0.3)

axes[-1].set_xlabel("Time (s)")
plt.tight_layout()
plt.savefig("single_trial_signals.png", dpi=150)
plt.show()

# 5b. 画各类别的平均 ERP（对通道取平均）
fig, ax = plt.subplots(figsize=(10, 5))
labels = np.unique(y)
colors = ['#e74c3c', '#3498db', '#2ecc71', '#f39c12']

for label, color in zip(labels, colors):
    mask = y == label
    avg_erp = X[mask].mean(axis=(0, 1))  # 对 trials 和 channels 取平均
    ax.plot(t, avg_erp, label=label, color=color, linewidth=1.2)

ax.set_xlabel("Time (s)")
ax.set_ylabel("Amplitude (µV)")
ax.set_title("Average ERP per Class (all channels)")
ax.legend()
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig("average_erp.png", dpi=150)
plt.show()
````
